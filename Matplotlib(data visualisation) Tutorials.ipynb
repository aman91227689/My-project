```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show() # here show function 
```

The syntax of the show() function is straightforward. It doesn't require any arguments. In this code snippet, we first import the pyplot module of the Matplotlib library and create a plot using the plot() function. Then we use the show() function to display the plot.28-Sept-2021



```python
plt.bar(x,y) # using bar function ie bar(), i can visualise the data in most efficient way
```

bar() function allows you to specify a starting value for a bar. Instead of running from zero to a value, it will go from the bottom to the value. The first call to pyplot. bar() plots the blue bars.

```python
import matplotlib.pyplot as plt
a=[1,2,3]
b=[2,4,6]
c=['r','g','y']
plt.bar(a,b,color=c)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
plt.bar(x,y,color=['r','b','g','y'])
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
plt.xlabel('Languages',fontsize=34)
plt.ylabel('Number')
plt.title('Language popularity')
plt.bar(x,y,color=['r','b','g','y'],width=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
plt.xlabel('Languages',fontsize=34)
plt.ylabel('Number')
plt.title('Language popularity')
plt.bar(x,y,color=['r','b','g','y'],width=0.3,align='edge') # inside align parameter there are two types of value ie edge and center
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
plt.xlabel('Languages',fontsize=34)
plt.ylabel('Number')
plt.title('Language popularity')
plt.bar(x,y,color=['r','b','g','y'],width=0.3,align='edge',edgecolor='m',linewidth=5,linestyle=':') # inside align parameter there are two types of value ie edge and center
plt.show()
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
plt.xlabel('Languages',fontsize=34)
plt.ylabel('Number')
plt.title('Language popularity')
plt.bar(x,y,color=['r','b','g','y'],width=0.3,align='edge',edgecolor='m',linewidth=5,linestyle=':',alpha=0.5) # inside align parameter there are two types of value ie edge and center
plt.show() # here alpha parameter is used to do light color 
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
plt.xlabel('Languages',fontsize=34)
plt.ylabel('Number')
plt.title('Language popularity')
plt.bar(x,y,color=['r','b','g','y'],width=0.3,align='edge',edgecolor='m',linewidth=5,linestyle=':',alpha=0.5,label='python')
plt.legend()# inside align parameter there are two types of value ie edge and center
plt.show() # here alpha parameter is used to do light color
 # here using legend() parameter i can put label in our data visualisation
```

```python
import matplotlib.pyplot as plt
x=['python','c','c++','java']
y=[85,75,70,82]
z=[20,30,40,50]
plt.xlabel('Languages',fontsize=34)
plt.ylabel('Number')
plt.title('Language popularity')
plt.bar(x,y,color='r',width=0.3,align='edge',linewidth=5,label='python')
plt.bar(x,z,color='y',width=0.3,align='edge',linewidth=5,label='python old')

plt.legend()# inside align parameter there are two types of value ie edge and center
plt.show() # here alpha parameter is used to do light color
 # here using legend() parameter i can put label in our data visualisation
```
